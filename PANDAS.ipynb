# seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize']=6,2

import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
vis1 = sns.distplot(df['InternetUsers'])
plt.show()
```

```python
vis1 = sns.distplot(df['InternetUsers'],bins = 15)
plt.show(vis1)
```

```python
vis2 = sns.distplot(df['InternetUsers'])
plt.show(vis1)
```

```python
vis4 = sns.boxplot(data = df, x = "IncomeGroup",y='BirthRate')
plt.show(vis4)
```

```python
vis5 = sns.lmplot(data = df, x = "InternetUsers",y='BirthRate')
plt.show(vis5)
```

```python
vis6 = sns.lmplot(data = df, x = "InternetUsers",y='BirthRate',fit_reg=False)
plt.show(vis6)  # Don't see the line when we set it false
```

```python
vis7 = sns.lmplot(data = df, x = "InternetUsers",y='BirthRate',fit_reg = False,hue="IncomeGroup")
plt.show(vis7) # hue is for colour information 
```

```python
vis8 = sns.lmplot(data = df, x = "InternetUsers",y='BirthRate',fit_reg = True,hue="IncomeGroup")
plt.show(vis8) # hue is for colour information 
```
